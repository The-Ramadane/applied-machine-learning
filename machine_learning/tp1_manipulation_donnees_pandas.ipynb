# Modification et assignation

```python
# 1 - Créer une nouvelle colonne appelée Rentabilité = BoxOffice / Budget, arrondie à 2 décimales.
df["Rentabilité"] = (df["BoxOffice_M€"] / df["Budget_M€"]).round(2)
df
```

```python
# 2 - Ajouter une colonne booléenne “Succès” qui vaut True si la rentabilité est > 2, sinon False.
df["Succès"] = df["Rentabilité"] > 2
df
```

```python
# 3 - Créez une nouvelle colonne "NoteNorm" correspondant à la note normalisée entre 0 et 1
df.loc[:, "NoteNorm"] = (df["Note"] - df["Note"].min()) / (df["Note"].max() - df["Note"].min())
df
```

```python
#  la rentabilité moyenne de chaque genre pour chaque pays et affichez uniquement les combinaisons ayant plus de 200 films.
stats = df.groupby(['Genre', 'Pays'])['Rentabilité'].agg(['mean', 'count'])

result = stats[stats['count'] > 200]
result
```

# Gestion des données manquantes

```python
# 1 - Identifier le nombre de valeurs manquantes par colonne.
df.isnull().sum()
```

```python
# 2 - Remplacer les valeurs manquantes de la colonne Note par la moyenne des notes.
df["Note"] = df["Note"].fillna(df["Note"].mean())
df
```

```python
# 3 - Remplacer les valeurs manquantes de la colonne Budget par la médiane du budget.
df["Budget_M€"] = df["Budget_M€"].fillna(df["Budget_M€"].median())
df
```

```python
# 4 - Supprimer les lignes contenant encore des valeurs manquantes (s’il en reste).
df = df.dropna()
df
```

```python
# 5 - Vérifier que le DataFrame ne contient plus de NaN.
df.isnull().sum()
```

# Agrégation et regroupement

```python
# 1 - Afficher la note moyenne par genre.
df.groupby("Genre")["Note"].mean()
```

```python
# 2 - Afficher le budget moyen et le BoxOffice moyen par réalisateur.
# df.groupby("Réalisateur").agg({"Budget_M€": "mean", "BoxOffice_M€": "mean"})
df.groupby('Réalisateur')[['Budget_M€','BoxOffice_M€']].agg(['mean','count'])
```

```python
# 3 - Trouver le film le plus rentable de chaque réalisateur.
df.groupby("Réalisateur")["Rentabilité"].idxmax()
```

```python
# 4- Calculez la rentabilité moyenne (Rentabilite) de chaque genre pour chaque pays. Affichez uniquement les combinaisons ayant plus de 200 films.
rentabilité_moyenne = df.groupby(["Genre", "Pays"])["Rentabilité"].agg(["mean", "count"])
rentabilité_moyenne[rentabilité_moyenne["count"] > 200]
```

# 3 - Pivot et réorganisation

```python
# 1 - Créer une nouvelle DataFrame dont :
# les lignes sont les Genre,
# les colonnes sont les Années,
# les valeurs sont les notes moyennes.
pivot_df = df.pivot_table(index="Genre", columns="Année", values="Note", aggfunc="mean")
pivot_df
```

```python
# 2 - Créer une nouvelle DataFrame pour comparer la rentabilité moyenne des genres au fil des années.
pivot_df_rentabilité = df.pivot_table(index="Genre", columns="Année", values="Rentabilité", aggfunc="mean")
pivot_df_rentabilité
```

# Questions bonus

```python
# 1 - Quels réalisateurs ont réalisé plus de 100 films ?
df['Réalisateur'].value_counts()[df['Réalisateur'].value_counts() > 100]
```

```python
# 2 - Quel est le genre qui a la rentabilité moyenne la plus élevée ?
df.groupby('Genre')['Rentabilité'].mean().idxmax()
```

```python
# 3 - Quelle est la durée moyenne des films de C. Morel ?
df[df['Réalisateur'] == 'L. Roy']["Durée_min"].mean()
```

```python
# 4 - Quels sont les 3 réalisateurs dont les films ont la plus faible variation de note (écart-type minimum) ? Ne pas inclure les réalisateurs avec moins de 3 films.
df.groupby('Réalisateur')["Note"].agg(["count", "mean", "std"]).query("count >= 3").sort_values('std', ascending=True)
```

```python
# 5 - Quels sont les films qui ont surperformés ? C'est-à-dire les films dont le BoxOffice est supérieur de plus de 50 % à la moyenne des films du même genre ?
box_office_moy_genre = df.groupby("Genre")["BoxOffice_M€"].transform("mean")
df.loc[:, "Ecart_Percent"] = ((df["BoxOffice_M€"] - box_office_moy_genre) / box_office_moy_genre) * 100
df[df["Ecart_Percent"] > 50][["Titre", "Réalisateur", "Ecart_Percent"]].sort_values(by="Ecart_Percent", ascending=False)
```
